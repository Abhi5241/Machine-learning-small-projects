Outlier remove stretagy

```python
def removeOutliers( col):
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    lb=Q1- 1.5* IQR
    ub=Q3+ 1.5* IQR
    return lb,ub
```

```python
for i in num:
    lb,ub=removeOutliers(i)
    df[i]=np.where(df[i] >ub,ub,df[i])
    df[i]=np.where(df[i] <lb,lb,df[i])
    
```

```python
for i in num:
    ax=sns.boxplot(data=df,x=i)
    ax.set(title=f"{i} Distribution",xlabel=f"{i} charge",ylabel="No of Customers")
    plt.show()
```

"""Hierarchical cluster"""

```python
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
df.drop("state",axis=1,inplace=True)
```

```python
final_df=df.iloc[:,:-1]
```

```python
output_df=df.iloc[:,-1]
```

```python
final_df.replace(to_replace=["no","yes"],value=[0,1],inplace = True)
```

```python
final_df.drop('phone number',axis=1,inplace=True)
```

```python
final_df
```

```python
warlink=linkage(final_df,method="ward")
```

```python
plt.figure(figsize=(12,8))
dendrogram(warlink,truncate_mode='lastp',p=10);
```

```python
from scipy.cluster.hierarchy import fcluster
```

```python
hie_clus=fcluster(warlink,3,criterion="maxclust")
```

```python
hie_clus
```

```python
final_df['Hie_Clus']=hie_clus
```

```python
final_df
```

```python
final_df.pivot_table(values=num,index=['Hie_Clus'],aggfunc="mean").T
```

```python
final_df.drop("Hie_Clus",axis=1,inplace=True)
```

Kmeans clustering

```python
from sklearn.preprocessing import StandardScaler
```

```python
x=StandardScaler()
```

```python
final_df_sc=x.fit_transform(final_df)
```

```python
final_df_sc
```

```python
from sklearn.cluster import KMeans,DBSCAN
```

```python
k_mean=KMeans(n_clusters=2)
k_mean.fit(final_df_sc)
```

```python
print(k_mean.labels_)
print(k_mean.inertia_)
```

```python
k_mean=KMeans(n_clusters=3)
k_mean.fit(final_df_sc)
print(k_mean.labels_)
print(k_mean.inertia_)
```

```python
k_mean=KMeans(n_clusters=10)
k_mean.fit(final_df_sc)
print(k_mean.labels_)
print(k_mean.inertia_)
```

```python
inertia=[]
for i in range(1,20):
    k_mean=KMeans(n_clusters=i)
    k_mean.fit(final_df_sc)
    inertia.append(k_mean.inertia_)
```
