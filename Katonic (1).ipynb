```python
x=df.drop(columns="Priority")
y=df["Priority"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x.sample()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder,LabelEncoder
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score

from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,VotingClassifier
```

# RANDOM FOREST


```python
step1=ColumnTransformer([
    ("tnf1",OrdinalEncoder(categories=[["High","Low","Not Reassign"]]),["# Reassignments"]),
    ("tnf2",OrdinalEncoder(categories=[["High interacted","Low Interacted"]]),["# Related Interactions"]),
    ("tnf3",OrdinalEncoder(categories=[["Urgently","Fast","medium","Slow"]]),["Timetaken"]),
    ("tnf4",OneHotEncoder(sparse=False,drop="first"),[0,1,2,3,5,7,8])
],remainder="passthrough")

step2=RandomForestClassifier(n_estimators=500,random_state=3,max_samples=0.5,max_features=0.75,max_depth=10)

pipe=Pipeline([
    ("step1",step1),
    ("step2",step2)
])

pipe.fit(x_train,y_train)

y_pred=pipe.predict(x_test)

print("Accuracy score:",accuracy_score(y_test,y_pred))
```

# GRADIENT BOOSTING

```python
step1=ColumnTransformer([
    ("tnf1",OrdinalEncoder(categories=[["High","Low","Not Reassign"]]),["# Reassignments"]),
    ("tnf2",OrdinalEncoder(categories=[["High interacted","Low Interacted"]]),["# Related Interactions"]),
    ("tnf3",OrdinalEncoder(categories=[["Urgently","Fast","medium","Slow"]]),["Timetaken"]),
    ("tnf4",OneHotEncoder(sparse=False,drop="first"),[0,1,2,3,5,7,8])
],remainder="passthrough")

step2=GradientBoostingClassifier(n_estimators=500)

pipe=Pipeline([
    ("step1",step1),
    ("step2",step2)
])

pipe.fit(x_train,y_train)

y_pred=pipe.predict(x_test)

print("Accuracy score:",accuracy_score(y_test,y_pred))
```

# SVM

```python
step1=ColumnTransformer([
    ("tnf1",OrdinalEncoder(categories=[["High","Low","Not Reassign"]]),["# Reassignments"]),
    ("tnf2",OrdinalEncoder(categories=[["High interacted","Low Interacted"]]),["# Related Interactions"]),
    ("tnf3",OrdinalEncoder(categories=[["Urgently","Fast","medium","Slow"]]),["Timetaken"]),
    ("tnf4",OneHotEncoder(sparse=False,drop="first"),[0,1,2,3,5,7,8])
],remainder="passthrough")

step2=SVC(kernel="rbf")

pipe=Pipeline([
    ("step1",step1),
    ("step2",step2)
])

pipe.fit(x_train,y_train)

y_pred=pipe.predict(x_test)

print("Accuracy score:",accuracy_score(y_test,y_pred))
```

# Voting 

```python
step1=ColumnTransformer([
    ("tnf1",OrdinalEncoder(categories=[["High","Low","Not Reassign"]]),["# Reassignments"]),
    ("tnf2",OrdinalEncoder(categories=[["High interacted","Low Interacted"]]),["# Related Interactions"]),
    ("tnf3",OrdinalEncoder(categories=[["Urgently","Fast","medium","Slow"]]),["Timetaken"]),
    ("tnf4",OneHotEncoder(sparse=False,drop="first"),[0,1,2,3,5,7,8])
],remainder="passthrough")

rf=RandomForestClassifier(n_estimators=500,random_state=3,max_samples=0.5,max_features=0.75,max_depth=10)
gbt=GradientBoostingClassifier(n_estimators=500)
ss=SVC(kernel="rbf")

step2=VotingClassifier([("rf",rf),("gbt",gbt),("ss",ss)])


pipe=Pipeline([
    ("step1",step1),
    ("step2",step2)
])

pipe.fit(x_train,y_train)

y_pred=pipe.predict(x_test)

print("Accuracy score:",accuracy_score(y_test,y_pred))
```

```python
print("Precision - ",precision_score(y_test,y_pred,average='weighted'))
print("Recall - ",recall_score(y_test,y_pred,average='weighted'))
print("F1 score - ",f1_score(y_test,y_pred,average='weighted'))
```
